```r
### Run this cell before continuing.
library(tidyverse)
library(repr)
library(tidymodels)
options(repr.matrix.max.rows = 6)
```

```r
heart_disease<- read_csv(file = "https://raw.githubusercontent.com/Chr1zz-D/dsci-100-demo/main/heart-disease.csv")
heart_disease <- mutate(heart_disease, cp= as_factor(cp),restecg=as_factor(restecg),slope=as_factor(slope),
                       exang=as_factor(exang))
heart_disease_scaled<-heart_disease |>
mutate (scaled_chol= scale(chol, center= TRUE),
       scaled_oldpeak= scale(oldpeak, center=TRUE),
       scaled_thalach=scale(thalach, center= TRUE),
       scaled_trestbps= scale(trestbps, center=TRUE))
heart_disease_scaled
```

```r
thalach_vs_chol_plot_slope <- ggplot(heart_disease_scaled, aes(x=scaled_chol, y=scaled_thalach,color=slope, shape=slope)) + geom_point() +
labs(x="serum cholestoral in mg/dl",y="maximum heart rate achieved", shape="the slope of the peak exercise ST segment",color="the slope of the peak exercise ST segment")
thalach_vs_chol_plot_slope
thalach_vs_chol_plot_cp <- ggplot(heart_disease_scaled, aes(x=scaled_chol, y=scaled_thalach,color=cp, shape=cp)) + geom_point() +
labs(x="serum cholestoral in mg/dl",y="maximum heart rate achieved", shape="chest pain type",color="chest pain type")
thalach_vs_chol_plot_cp
thalach_vs_chol_plot_restecg <- ggplot(heart_disease_scaled, aes(x=scaled_chol, y=scaled_thalach,color=restecg, shape=restecg)) + geom_point() +
labs(x="serum cholestoral in mg/dl",y="maximum heart rate achieved", shape=" resting electrocardiographic results",color=" resting electrocardiographic results")
thalach_vs_chol_plot_restecg
thalach_vs_chol_plot_exang <- ggplot(heart_disease_scaled, aes(x=scaled_chol, y=scaled_thalach,color=exang, shape=exang)) + geom_point() +
labs(x="serum cholestoral in mg/dl",y="maximum heart rate achieved", shape="exercise induced angina (1 = yes; 0 = no)",color="exercise induced angina (1 = yes; 0 = no)")
thalach_vs_chol_plot_exang
```

```r
heart_disease_split <- initial_split(heart_disease, prop = 0.5, strata = cp)  
heart_disease_train <- training(heart_disease_split)   
heart_disease_test <- testing(heart_disease_split) 
```

```r
knn_tune <-  nearest_neighbor(weight_func = "rectangular", neighbors = tune())|>
       set_engine("kknn") |>
       set_mode(mode = "classification")
```

```r
heart_disease_vfold <- vfold_cv(heart_disease_train, v =  5, strata= slope)
```

```r
heart_disease_recipe <- recipe(exang ~chol+thalach, data=heart_disease_train)  |>
 step_center(all_predictors())|>
step_center(all_predictors())
```

```r
knn_results <- workflow()|>
add_recipe(heart_disease_recipe)|>
add_model(knn_tune) |>
tune_grid(resamples= heart_disease_vfold, grid= 10)|>
collect_metrics()
knn_results
```

```r
accuracies <- knn_results |>
filter(.metric == "accuracy")
accuracy_versus_k <- ggplot(accuracies, aes(x = neighbors, y = mean))+
       geom_point() +
       geom_line() +
       labs(x = "Neighbors", y = "Accuracy Estimate") +
       scale_x_continuous(breaks = seq(0, 14, by = 1)) +  
       scale_y_continuous(limits = c(0.4, 1.0)) 
accuracy_versus_k
```

```r
heart_disease_knn <- nearest_neighbor(weight_func = "rectangular", neighbors = 15)|>
       set_engine("kknn") |>
       set_mode(mode = "classification")
```

```r
hear_disease_fit <- workflow() |>add_recipe ( heart_disease_recipe) |>
add_model( heart_disease_knn) |>
fit(data= heart_disease_train)
```

```r
hear_disease_prediction <- predict(hear_disease_fit,heart_disease_test) |> bind_cols(heart_disease_test)|>select(.pred_class, restecg, chol, thalach)
hear_disease_prediction
```
