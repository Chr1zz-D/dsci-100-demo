# Exploratory Data Analysis and Predictive Modeling of Heart Disease Indicators

### Introduction 
Heart disease, a broad term for a variety of cardiovascular conditions, continues to be a leading cause of death worldwide. It is a multi-factorial disorder influenced by both genetic and lifestyle factors, among others. Early prediction and detection of heart disease are key to improving patient outcomes, reducing the risk of severe complications, and ensuring effective treatment.

Given this context, the primary objective of our project was to identify key predictors of exercise-induced angina, a symptom associated with heart disease. We used visulization method to ensure the heart disease type we explore. Exercise-induced angina is chest pain that occurs due to reduced blood flow to the heart muscles, typically triggered by physical exertion. By identifying the key predictors of exercise-induced angina, we hoped to shed light on potential early indicators of heart disease, enabling timely intervention and treatment.

To answer this question, we employed a dataset obtained from UCI Machine Learning Repository (Janosi et al., 1988). This dataset comprises various heart health indicators such as serum cholesterol levels (cholesterol content in the blood), maximum heart rate achieved (the highest heart rate an individual can reach), resting blood pressure, fasting blood sugar, resting EKG results, age, sex, among others (Joy, M et al., 1982). 

In the next sections, we will explain the data cleaning and pre-processing steps, present our exploratory data analysis, detailed building our predictive model, and finally, discuss the findings and their potential implications.

### Methods
1. Import data:
  The dataset "heart-disease.csv" is imported using the read_csv() function in the tidyverse library. The source of data we used is from a free website and stored in our github project (Janosi,et al, 1988). The data is stored in the variable heart_disease.

2. Converting categorical variables:
  For categorical variables of obvious nature, the variables we used are "cp" (type of chest pain), "restecg" (resting electrocardiographic results), "slope" (slope of peak ST segment of exercise) and "exang" (exercise induced angina) (Janosi,et al, 1988). These variables are converted to factors using the as_factor() function. This conversion allows to treat these variables as categories rather than continuous values, facilitating subsequent analysis and modeling.

3. Scaling variables:
  The variables used as the predictors are "chol" (serum cholestoral in mg/dl), "oldpeak" (ST depression induced by exercise relative to rest), "thalach" (maximum heart rate achieved) and "trestbps" (resting blood pressure at admission).(Janosi,et al, 1988) They were scaled using the scale() function for scaling. The scaled variables are stored in a new column prefixed with "scaled_" in the heart_disease_scaled data frame. Scaling ensures that these variables are on a similar scale, avoiding any bias in subsequent analyses and allowing for meaningful comparisons between variables.

4. Making scatterplot to fin the related variables:
  Visualizations were generated to understand the relationships between these predictors and other factors such as 'slope of the peak exercise ST segment', 'chest pain type', 'resting electrocardiographic results', and 'exercise-induced angina'. By comparing the plots, the variable 'exercise-induced angina' was chosen as the object to predict. Then a plot of oldpeak vs trestbps was generated to make sure that serum cholestoral and maximum heart rate achieved were most related variables.


5. Build model:
A K-Nearest Neighbors (KNN) classification model was built to predict the presence of exercise-induced angina using serum cholesterol and maximum heart rate as predictors (de Divitiis, et al., 1996). The model makes predictions based on the similarity between new data points and the labeled data points in the training set. Therefore, training set and testing set were formed with the proportion of 0.7. To find the most suitable number of the neighbours, the vfold data were generated. 5 vfolds were based on the training set and the number of neighbours were from 1 to 15 with the difference of 1. Then the metrix were collected and a plot of accuracies rate was built which is better to find the best number of neighbours. The optimal number of neighbors (n=4) was identified via cross-validation. After that, the completed model was generated based on the neighbour of 4 and made the prediction of the testing set.
  
### Data Analysis

```r
### Run this cell before continuing.
library(tidyverse)
library(repr)
library(tidymodels)
options(repr.matrix.max.rows = 6)
```

```r
heart_disease<- read_csv(file = "https://raw.githubusercontent.com/Chr1zz-D/dsci-100-demo/main/heart-disease.csv")
heart_disease <- mutate(heart_disease, cp= as_factor(cp),restecg=as_factor(restecg),slope=as_factor(slope),
                       exang=as_factor(exang))
heart_disease_scaled<-heart_disease |>
mutate (scaled_chol= scale(chol, center= TRUE),
       scaled_oldpeak= scale(oldpeak, center=TRUE),
       scaled_thalach=scale(thalach, center= TRUE),
       scaled_trestbps= scale(trestbps, center=TRUE))
heart_disease_scaled
```

```r
thalach_vs_chol_plot_slope <- ggplot(heart_disease_scaled, aes(x=scaled_chol, y=scaled_thalach,color=slope, shape=slope)) + geom_point() +
labs(x="serum cholestoral in mg/dl",y="maximum heart rate achieved", shape="the slope of the peak exercise ST segment",color="the slope of the peak exercise ST segment")
thalach_vs_chol_plot_slope
thalach_vs_chol_plot_cp <- ggplot(heart_disease_scaled, aes(x=scaled_chol, y=scaled_thalach,color=cp, shape=cp)) + geom_point() +
labs(x="serum cholestoral in mg/dl",y="maximum heart rate achieved", shape="chest pain type",color="chest pain type")
thalach_vs_chol_plot_cp
thalach_vs_chol_plot_restecg <- ggplot(heart_disease_scaled, aes(x=scaled_chol, y=scaled_thalach,color=restecg, shape=restecg)) + geom_point() +
labs(x="serum cholestoral in mg/dl",y="maximum heart rate achieved", shape=" resting electrocardiographic results",color=" resting electrocardiographic results")
thalach_vs_chol_plot_restecg
thalach_vs_chol_plot_exang <- ggplot(heart_disease_scaled, aes(x=scaled_chol, y=scaled_thalach,color=exang, shape=exang)) + geom_point() +
labs(x="serum cholestoral in mg/dl",y="maximum heart rate achieved", shape="exercise induced angina (1 = yes; 0 = no)",color="exercise induced angina (1 = yes; 0 = no)")
thalach_vs_chol_plot_exang 
oldpeak_vs_trestbps_exange <-ggplot(heart_disease_scaled, aes(x=scaled_oldpeak, y=scaled_trestbps,color=exang, shape=exang)) + geom_point() +
labs(x="ST depression induced by exercise relative to rest", y="resting blood pressure (in mm Hg on admission to the hospital)",
     shape="exercise induced angina (1 = yes; 0 = no)",color="exercise induced angina (1 = yes; 0 = no)")
oldpeak_vs_trestbps_exange
```

```r

set.seed(2023)

heart_disease_split <- initial_split(heart_disease, prop = 0.7, strata = cp)  
heart_disease_train <- training(heart_disease_split)   
heart_disease_test <- testing(heart_disease_split) 
```

```r
set.seed(2023)


knn_tune <-  nearest_neighbor(weight_func = "rectangular", neighbors = tune())|>
       set_engine("kknn") |>
       set_mode(mode = "classification")
```

```r
set.seed(2023)

heart_disease_vfold <- vfold_cv(heart_disease_train, v =  5, strata= slope)
```

```r
set.seed(2023)

heart_disease_recipe <- recipe(exang ~chol+thalach, data=heart_disease_train)  |>
 step_center(all_predictors())|>
step_center(all_predictors())
```

```r
set.seed(2023)


knn_results <- workflow() |>
add_recipe(heart_disease_recipe) |>
add_model(knn_tune) |>
tune_grid(resamples = heart_disease_vfold, grid= 10)|>
collect_metrics()
knn_results
```

```r
set.seed(2023)

accuracies <- knn_results |>
filter(.metric == "accuracy")
accuracy_versus_k <- ggplot(accuracies, aes(x = neighbors, y = mean))+
       geom_point() +
       geom_line() +
       labs(x = "Neighbors", y = "Accuracy Estimate") +
       scale_x_continuous(breaks = seq(1, 15, by = 1)) +  
       scale_y_continuous(limits = c(0.4, 1.0)) 
accuracy_versus_k

accuracy_results <- knn_results |>
  filter(.metric == "accuracy")

print(accuracy_results)
```

```r
set.seed(2023)

heart_disease_knn <- nearest_neighbor(weight_func = "rectangular", neighbors = 4) |>
       set_engine("kknn") |>
       set_mode(mode = "classification")
```

```r
set.seed(2023)


hear_disease_fit <- workflow() |> add_recipe ( heart_disease_recipe) |>
add_model(heart_disease_knn) |>
fit(data = heart_disease_train)
```

```r
hear_disease_prediction <- predict(hear_disease_fit,heart_disease_test) |> bind_cols(heart_disease_test)|> select(.pred_class, restecg, chol, thalach)
hear_disease_prediction
```

### Results

The model was used to make predictions on the test data. The prediction results were joined with the test set data, and the predictions along with the original restecg, chol, and thalach variables were selected for a final output. These prediction results can be utilized to understand the likelihood of exercise-induced angina given the levels of 'serum cholesterol' and 'maximum heart rate achieved' in future patients.

### Discussion

The data analysis showed some meaningful patterns in the relationships between cholesterol, heart rate, and other factors such as the 'slope of the peak exercise ST segment', 'chest pain type', and 'resting electrocardiographic results'. We then applied the k-nearest neighbors (KNN) algorithm to the data and found that the optimal number of neighbors for accurate prediction was 4.The resulting model yielded an accuracy of approximately 0.744. Although we anticipated that cholesterol and heart rate would have a significant impact on the occurrence of exercise-induced angina, the accuracy rate of 0.744  was not as high as we initially hoped. This suggests that while these predictors are important, other factors not considered in this study may also significantly influence the onset of exercise-induced angina.

The findings of this study could potentially be used to inform medical professionals about the risk factors for exercise-induced angina in patients with heart disease. This could enable early intervention strategies and personalized treatment plans to manage this condition more effectively.
However, as the accuracy of the model was not exceptionally high, caution must be exercised in its application. Future research could incorporate additional predictors such as age, sex, and lifestyle factors to improve the predictive accuracy. In addition, more data samples could be employed to handle the complexity of this health issue.
Furthermore, this study may spark further research into understanding the relationship between exercise-induced angina and other physiological factors (de Divitiis, et al., 1996). For instance, how does the interplay between cholesterol levels, maximum heart rate, and other potential predictors influence the risk of exercise-induced angina? How does this relationship change across different patient populations? These questions and others could provide a direction for future investigations.
